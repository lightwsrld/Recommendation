# 리뷰 크롤링 (전체 페이지)

```python
driver = webdriver.Chrome()
driver.get("https://tickets.interpark.com/goods/P0003343")
driver.implicitly_wait(5) # HTML
time.sleep(2)
try:
    close_button = driver.find_element(By.CLASS_NAME, "popupCloseBtn.is-bottomBtn")
    close_button.click()
    time.sleep(0.5)
except NoSuchElementException:
    pass

    time.sleep(1)
nav = driver.find_elements(By.CLASS_NAME, "navItem ")
for i in nav:
    if "관람후기" in i.text:
        i.click()
        time.sleep(2)
total = []
while True:
    pages = driver.find_element(By.CLASS_NAME, "pageNumWrap").find_elements(By.TAG_NAME, "li")
    for i in pages:
        i.click()
        time.sleep(2)
        for i in driver.find_elements(By.CLASS_NAME, "bbsItem"):
            title = i.find_element(By.CLASS_NAME, "bbsTitle").text
            content = i.find_element(By.CLASS_NAME, "bbsText").text
            total.append([title, content])
        print(f"{len(total)}개의 리뷰를 저장했습니다.")

    try:
        driver.find_element(By.CLASS_NAME, "pageNextBtn.pageArrow").click()
    except:
        break
```

```python
# 500개 이상 리뷰 크롤링

driver = webdriver.Chrome()
driver.get("https://tickets.interpark.com/goods/23006752")
driver.implicitly_wait(5) # HTML
time.sleep(3)
driver.find_element(By.CLASS_NAME, "popupCloseBtn.is-bottomBtn").click()
time.sleep(1)
nav = driver.find_elements(By.CLASS_NAME, "navItem ")
for i in nav:
    if "이용후기" in i.text:
        i.click()
        time.sleep(5)
total = []
while True:
    pages = driver.find_element(By.CLASS_NAME, "pageNumWrap").find_elements(By.TAG_NAME, "li")
    for i in pages:
        i.click()
        time.sleep(2)
        for i in driver.find_elements(By.CLASS_NAME, "bbsItem"):
            title = i.find_element(By.CLASS_NAME, "bbsTitle").text
            content = i.find_element(By.CLASS_NAME, "bbsText").text
            total.append([title, content])
        print(f"{len(total)}개의 리뷰를 저장했습니다.")

    try:
        pos = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CLASS_NAME, "pageNextBtn.pageArrow")))
        pos.click()
    except:
        break

driver.quit()
```

```python
total_1 = pd.DataFrame(total, columns=['제목', '리뷰'])
total_1.head()
```

```python
total_1.tail()
```

```python
# 1000개 이상 리뷰 크롤링

driver = webdriver.Chrome()
driver.get("https://tickets.interpark.com/goods/21002623")
driver.implicitly_wait(5) # HTML
time.sleep(3)
driver.find_element(By.CLASS_NAME, "popupCloseBtn.is-bottomBtn").click()
time.sleep(1)
nav = driver.find_elements(By.CLASS_NAME, "navItem ")
for i in nav:
    if "관람후기" in i.text:
        i.click()
        time.sleep(5)
total_운빨로맨스 = []
while True:
    pages = driver.find_element(By.CLASS_NAME, "pageNumWrap").find_elements(By.TAG_NAME, "li")
    for i in pages:
        i.click()
        time.sleep(2)
        for i in driver.find_elements(By.CLASS_NAME, "bbsItem"):
            title = i.find_element(By.CLASS_NAME, "bbsTitle").text
            content = i.find_element(By.CLASS_NAME, "bbsText").text
            total_운빨로맨스.append([title, content])
        print(f"{len(total_운빨로맨스)}개의 리뷰를 저장했습니다.")
    # 예외처리: 만약 try 안에서 에러가 발생하면 에러를 발생시키지않고, except를 실행
    try:
        pos = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CLASS_NAME, "pageNextBtn.pageArrow")))
        pos.click()
    except:
        break

driver.quit()
```

```python
total_운빨로맨스 = pd.DataFrame(total_운빨로맨스, columns=['제목', '리뷰'])
total_운빨로맨스.head()
```
